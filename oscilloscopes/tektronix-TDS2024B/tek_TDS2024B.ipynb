## Example of how to control Tektronix Oscilloscope TDS2024B using PyVISA 

```python
from osctck import *
```

```python
TDS2024 = Osctck('USB0::1689::874::C034414::0::INSTR')
```

```python
TDS2024.config(channels = (1,),
                 triggerSource = 'EXT',
                 triggerLevel = 0.5,
                 triggerSlope = 'RISE',
                 triggerMode = 'NORM',
                 triggerCoup = 'AC',
                 acquisition = 4,
                 vAutoScale = False,
                 chanband = 'OFF',
                 chaninv = 'OFF' )
```

```python
meas = TDS2024()
```

```python
import matplotlib.pyplot as plt
```

```python
t = meas[0]
v = meas[1] 
plt.figure()
plt.plot(t*1e6,v*1e3)
plt.xlabel('Time ($\mu$s)')
plt.ylabel('Amplitude (mV)')
plt.grid(linestyle = '--')
```
