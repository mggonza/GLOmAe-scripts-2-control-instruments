## Example of how to control Tektronix Oscilloscope TDS1012B using PyVISA 

```python
from osctck import *
```

```python
TDS1012B = Osctck('USB0::1689::870::C051633::0::INSTR')
```

```python
TDS1012B.config(channels = (1,),
                 triggerSource = 'EXT',
                 triggerLevel = 0.5,
                 triggerSlope = 'RISE',
                 triggerMode = 'NORM',
                 triggerCoup = 'AC',
                 acquisition = 4,
                 vAutoScale = False,
                 chanband = 'OFF',
                 chaninv = 'OFF' )
```

```python
meas = TDS1012B()
```

```python
import matplotlib.pyplot as plt
```

```python
t = meas[0]
v = meas[1] 
plt.figure()
plt.plot(t*1e6,v*1e3)
plt.xlabel('Time ($\mu$s)')
plt.ylabel('Amplitude (mV)')
plt.grid(linestyle = '--')
```
